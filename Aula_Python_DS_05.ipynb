```python
#################### ESTRUTURANDO O CÓDIGO #####################
# Librarys
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import plotly.express as px

# Function- Requisitos de uma funçao
  #1. Nome em relaçao a sua responsabilidade
  #2. Imputs_ Parametros de entrada
  #3. Output_ Dados de saìda

def show_dypes (fab):
    print(fab.dtypes)
    return None

def show_dimensions (fab):
    print('Number of rows:{}'.format(fab.shape[0]))
    print('Number of cols:{}'.format(fab.shape[1]))
    return None


def collect_geodata(fab,cols):

    #initialize API
    geolocator=Nominatim(user_agent='geoapiExercises')
    
    #create empty columns
    fab.loc[:,cols[0]]='NA'
    fab.loc[:,cols[1]]='NA'

    for i in range(len(fab)):
         print('Loop:{}/{}'.format(i,len(fab)))

         query=str(fab.loc[i,'lat'])+','+str(fab.loc[i,'long'])
         response=geolocator.reverse(query)

         if cols[0] in response.raw['address']:
            fab.loc[i,'house_number']=response.raw['address'][cols[0]]
         if cols[1] in response.raw['address']:
            fab.loc[i,'road']        =response.raw['address'][cols[0]]
    return fab


def fab_collect (path):
    # Load dataset
    fab=pd.read_csv('datasets/kc_house_data.csv')
    #fab=fab.head(20)#just sample

    # Extration analysis
    # Data dimensions
    show_dimensions(fab)

    ## Data Types
    show_dypes(fab)
    
    return fab

def df_trasform(fab):
        # Convert object to data
    fab['date']=pd.to_datetime(fab['date'])

    # Descriptive statistics
    num_attributes=fab.select_dtypes(include=('int64','float64'))

    # Central tendency
    pd.set_option('display.float_format', lambda x: f'%.{len(str(x%1))-2}f' % x) #tira a notaçao cientifica
    media=pd.DataFrame(num_attributes.apply(np.mean, axis=0))
    mediana=pd.DataFrame(num_attributes.apply(np.median,axis=0))

    # dispersion 
    maximo=pd.DataFrame(num_attributes.apply(np.max,axis=0))
    minimo=pd.DataFrame(num_attributes.apply(np.min,axis=0))
    std=pd.DataFrame(num_attributes.apply(np.std,axis=0))

    df1=pd.concat([maximo,minimo,media,mediana,std], axis=1).reset_index()
    df1.columns=['attributes','maximo','minimo','media','mediana','std']

    show_dimensions(df1)

    ### Crie uma nova coluna chamada 'dormitory_type'
    #- se o valor dacoluna 'bedrooms' for igual a 1 = studio
    #- se o valor dacoluna 'bedrooms' for igual a 2 = apartment
    #- se o valor dacoluna 'bedrooms' for igual a 3 = house

    fab ['dormitory_type'] = 'NA'

    for i in range(len(fab)):

        if fab.loc[i,'bedrooms'] ==1:
            fab.loc[i,'dormitory_type'] = 'Studio'

        elif fab.loc[i,'bedrooms'] ==2:
            fab.loc[i,'dormitory_type'] = 'Apartament'
        
        else:
            fab.loc[i,'dormitory_type'] = 'House'


    ### Qual a quantidade de imóveis por nível?
    #- Nivel 0: Preço entre R$ 0.00 e R$ 321.950
    #- Nivel 1: Preço entre R$ 321.950 e R$ 450.000
    #- Nivel 2: Preço entre R$ 450.000 e R$ 645.000
    #- Nivel 3: Preço entre R$ Acima de R$ 645.000

fab ['level'] = "NA"

for i in range ( len( fab ) ):
    if fab.loc[i, 'price'] <= 321950:
        fab.loc[i, 'level'] = 0

    elif (fab.loc[i, 'price'] > 321950) & (fab.loc [i,'price'] <= 450000):
        fab.loc[i, 'level'] = 1

    elif (fab.loc[i, 'price'] > 450000) & (fab.loc [i,'price'] <= 645000):
        fab.loc[i, 'level'] = 2

    elif (fab.loc[i, 'price'] > 645000) & (fab.loc [i,'price'] <= 950000):
        fab.loc[i, 'level'] = 3
        
    elif (fab.loc[i, 'price'] > 950000) & (fab.loc [i,'price'] <= 1500000):
        fab.loc[i, 'level'] = 4

    else:
        fab.loc[i, 'level'] = 5

    return fab

def df_load(fab):
            #cols=['road','house_number']
            #collect_geodata(df,cols)
            #map
    houses = fab [['id','long', 'lat', 'price','level']].copy()

    map = px.scatter_mapbox(houses,
                      lat='lat',
                      lon='long',
                      color='level',
                      size='price',
                      color_continuous_scale=px.colors.cyclical.IceFire,
                      size_max=15,
                      zoom=10)

    map.update_layout(mapbox_style='open-street-map')
    map.update_layout(height=300,margin={'r':0,'t':0,'l':0,'b':0})
    map.show()
    
    return None

if __name__=='__main__':

    #-------------------------------------
    # ETL
    #-------------------------------------

            # Collect
    fab_raw = fab_collect('kc_house_data')

            # Trasform

    fab_processing = fab_trasform(fab_raw)

            # Load
    fab_load(fab_processing)
```
